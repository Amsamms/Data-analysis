```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('../input/wili-2018/data.csv')
```

```python
df.info()
```

```python
df['class'].unique()
```

```python
dataset=df[df['class'].str.contains('eng|deu|spa|ita|fra')]
```

```python
X=df['text']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
training_data = count_vector.fit_transform(X)
```

```python
X_df=pd.DataFrame(data=training_data.toarray(),columns=count_vector.get_feature_names_out())
```

```python
X_df
```

```python
y=df['class']
```

```python
# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the 'features' and 'income' data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train, y_train)
```

```python
y_train_prediction=naive_bayes.predict(X_train)
y_test_prediction=naive_bayes.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('train accuracy ',accuracy_score(y_train,y_train_prediction))
print('test accuracy ',accuracy_score(y_test,y_test_prediction))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
pip=make_pipeline((CountVectorizer()),MultinomialNB())

```

```python
pip.fit(X_train,y_train)
```

```python
y_train_predict=pip.predict(X_train)
y_test_predict=pip.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('train accuracy ',accuracy_score(y_train,y_train_predict))
print('test accuracy ',accuracy_score(y_test,y_test_predict))
```
