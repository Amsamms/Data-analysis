```python
import numpy as np
import pandas as pd
from datetime import datetime
import pywhatkit
```

```python
# import google contacts from csv file and reset index
df=pd.read_csv('contacts.csv')
df.reset_index(inplace=True)
```

```python
# Turning the values of dataframe to a list
contacts_list=[]
for i in df.columns :
    for j in df.index:
        contacts_list.append(df.loc[j,i])
contacts_series=pd.Series(contacts_list)        
```

```python
# moving all numbers starting with + folloed by at least 3 digits in numpy series
contacts_numbers_startwithplus=contacts_series[contacts_series.str.contains("^\+\d{3,}",na=False,regex=True)].values
```

```python
# moving all numbers at least 5 digits in numpy series and then add +2 before it
contacts_numbers_startwithnumberonly=contacts_series[contacts_series.str.contains("^\d{5,}",na=False,regex=True)].values
contacts_numbers_startwithplus_2="+2"+ contacts_numbers_startwithnumberonly
```

```python
# concacenate the above two arrays
contacts_numbers=np.concatenate([contacts_numbers_startwithplus,contacts_numbers_startwithplus_2])
```

```python
# apply whatsapp function to all contacts list that have numbers, if error occured continue the loop
# using date time to determine the current hour and current minute

for number in contacts_numbers:
    try:
        pywhatkit.sendwhatmsg(number,"كل عام و أنتم بخير بمناسبة عيد الفطر المبارك\n احمد صبري", datetime.now().hour,datetime.now().minute+1
                              ,tab_close=True,close_time=6)
    except:
        continue
```
